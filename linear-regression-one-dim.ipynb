##Linear Regression
###1-dimensional

```python
%pylab inline
%matplotlib inline
```

```python
# original function
x = np.arange(-5,5,.1)
y = .2*x+3

# noise
mu = 0
sigma = 1
noise = sigma * randn(x.shape[0]) + mu

data = y+noise # 1 x n

order = 2 # includes static bias
dm = np.zeros((order,data.shape[0]))
for o in range(order):
    dm[o,:] = pow(x,o)

dmsq = dot(dm,transpose(dm))
dminv = inv(dmsq)
betav = dot(dot(dminv,dm),data)
print betav

prediction = zeros(x.shape)
for o in range(order):
    prediction += betav[o] * pow(x,o)
```

```python
# plots
figure(figsize=(16,16))
plot(x,y,c="b",label='Original')
plot(x,prediction,c='r',label='Prediction')
scatter(x,data,c='y',marker='x',label='Noisy data')
axes().set_aspect('equal')
legend(loc=0)
```
